# Download name.basics.tsv and Read

```python
import pandas as pd
actors = pd.read_csv('/Users/apillai2/Downloads/name.basics.tsv', sep='\t', header=0)
```

```python
print("Actor Count before Filter:",len(actors))
actors=actors[(actors.birthYear != '\\N')]
actors = actors[(actors.birthYear > '1950') & (actors.primaryProfession.str.contains('act'))]

actors['titleCount'] = actors['knownForTitles'].str.count(',')+1
print("Actor Count after Filter:",len(actors))
actors = actors[(actors.titleCount > 3)]
actors.head()
```

```python
import pandas as pd
princ = pd.read_csv('/Users/apillai2/Downloads/title.principals.tsv', sep='\t', header=0)
princ.head()
```

```python
princ[(princ.nconst == 'nm0000138') & (princ.category == 'actor')]
```

```python
print(len(actors))
princ_actors = pd.DataFrame(princ[(princ.category.str.contains('act'))]['nconst'].unique())
princ_actors.columns = ['nconst']
actors = actors.merge(princ_actors,how='inner')
del princ
```

```python
print(len(actors))
```

```python
import pandas as pd
titles = pd.read_csv('/Users/apillai2/Downloads/title.akas.tsv', sep='\t', header=0)
titles.head()
```

```python
df_titles = pd.DataFrame((titles[titles.region == 'US']['titleId'].unique()))
del titles
df_titles.columns = ['actor_prim_movie']
df_titles.head()
```

```python
import pandas as pd
ratings = pd.read_csv('/Users/apillai2/Downloads/title.ratings.tsv', sep='\t', header=0)
print(len(ratings))
ratings.head()
```

```python
ratings = ratings[ratings.numVotes > 1000]
print(len(ratings))
ratings = ratings[['tconst','numVotes']]
ratings.columns = ['actor_prim_movie','numVotes']
df_titles = df_titles.merge(ratings,how='inner')
actors['totnumVotes'] = 0
print(len(df_titles))
df_titles.head()
```

```python
def actor_prim_movie(actors,k):
    actor_prim_movie = []
    for i in actors['knownForTitles'].str.split(","):
        actor_prim_movie.append(i[k])
    return actor_prim_movie
##
for k in range(4):
    actors['actor_prim_movie'] = actor_prim_movie(actors,k)
    actors = actors.merge(df_titles,how='inner',on=['actor_prim_movie'])
    actors['totnumVotes'] = actors['totnumVotes'] + actors['numVotes']
    actors.drop(['numVotes'],axis=1,inplace=True)
    print(len(actors))
###
actors.head()
```

```python
actors = actors.sort_values('totnumVotes',ascending=False)
actors = actors.reset_index(drop=True)
actors.head()
```

# Write File

```python
actors.to_csv('actorsOrig.csv',header=True,index=False)
```
