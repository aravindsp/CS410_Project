```python
import pandas as pd
from rank_bm25 import *
import glob
import re
```

```python
actors=pd.read_csv('actors.csv') #.set_index('nconst')
actors.head()
```

# Get All Documents

```python
# All files ending with .txt
fileList = glob.glob("movieFile/*.txt")
```

```python
## 'movieFile/tt0443701.txt',
docList = []
movList = []
movText = []
for fl in fileList:
    with open(fl, 'r') as file:
        data = file.read().replace('\n', '')
        dataSamp = data[:500]
        movname = fl.replace('movieFile/','').replace('.txt','')
        data = movname+': '+data
        #data = spl_chars_removal(data)
        #data = stopwprds_removal_gensim_custom(data)
    docList.append(data)
    movList.append(movname)
    movText.append(dataSamp)
df = pd.DataFrame(movList)
df.columns = ['tconst']
df['mov_text'] = movText
df.head()
```

# Tokenise

```python
tokenized_corpus = [doc.split(" ") for doc in docList]
bm25 = BM25Okapi(tokenized_corpus)
```

# Test with Sample Query

```python
query = "time travel" ## Enter search query
tokenized_query = query.split(" ")
```

```python
#doc_scores = bm25.get_scores(tokenized_query)
#print(doc_scores)
rankedDocs = bm25.get_top_n(tokenized_query, docList, n=5)
##rankedDocs
```

```python
doc_scores = bm25.get_scores(tokenized_query)
df['doc_scores'] = doc_scores
df = df.sort_values('doc_scores',ascending=False)
df.head(100)
```

```python
df = df.merge(actors,how='inner')
df.head()
```

```python
df_actor = df.copy()
##df_actor['cum_score'] = df_actor.groupby('nconst')['doc_scores'].transform("sum")
#df_actor['cum_count'] = df_actor.groupby('nconst')['doc_scores'].transform("count")
df_actor['cum_rank'] = df_actor.groupby('nconst')['doc_scores'].rank("dense", ascending=False)
df_actor = df_actor[df_actor.cum_rank <= 4]
df_actor['cum_score'] = df_actor.groupby('nconst')['doc_scores'].transform("sum")
df_actor = df_actor.sort_values(['cum_score','doc_scores'],ascending=False)
df_actor = df_actor[:20].reset_index()
df_actor
```

```python
df_disp = df_actor.groupby(['nconst','primaryName','cum_score'])['tconst'].apply(list).reset_index()
df_disp_mov = df_actor.groupby(['nconst'])['mov_text'].apply(list).reset_index()
df_disp = df_disp.merge(df_disp_mov,how='inner').sort_values('cum_score',ascending=False).reset_index(drop=True)
##
df_disp.head()
```
